```python
import numpy as np
from matplotlib import patches 
import matplotlib.pyplot as plt

# Datos
x = ([2024, 2020, 2016, 2012, 2008, 2004, 2000, 1996, 1992, 1988, 1984, 1980])
y = (33.38, 32, 30.75, 33.33, 30.75, 30.25, 30, 27.67, 23.71, 27, 24.67, 26)
y2 = (29.23, 29, 27.38, 25.96, 27.31, 26.76, 25.6, 23.98, 21.27, 21.61, 20.86, 21.67)
y3 = (26.20, 25, 22, 19.20, 20.80, 21.67, 21.67, 20.17, 19, 17, 16, 14.67)

# Gráfico
fig, ax = plt.subplots(figsize=(35,15), facecolor='#001432')
ax.set_facecolor('#191c15')

# Grid
plt.grid(axis='y', alpha=1, linewidth=0.4, color='white', zorder=0)
plt.grid(axis='x', alpha=1, linewidth=0.4, color='white', zorder=0)

# Puntos
plt.scatter(x, y, marker='^', color='#bfff00', label='MAX', edgecolors = '#434547', linewidths = 5, s=1750, zorder=3)
plt.scatter(x, y2, marker='o', color='white', label='Average', edgecolors = '#434547', linewidths = 7, s=1750, zorder=3)
plt.scatter(x, y3, marker='v', color='#faad1e', label='MIN', edgecolors = '#434547', linewidths = 5, s=1750, zorder=3)

# Etiquetas de puntos
for i, label in enumerate(y):
    plt.annotate(label, (x[i], y[i] + 2), ha='center', va='bottom', fontsize=27.5, color='#c6ff00', weight='bold')
for i, label in enumerate(y3):
    plt.annotate(label, (x[i], y3[i] - 4), ha='center', va='bottom', fontsize=27.5, color='#faad1e', weight='bold')
    
# Ejes
plt.yticks(np.arange(0, 41, 5))
plt.xticks([2024, 2020, 2016, 2012, 2008, 2004, 2000, 1996, 1992, 1988, 1984, 1980])
plt.tick_params(labelsize=25)

# Label
ax.tick_params(colors='white', length=15, width=1.5, which='both')

# Bordes
for axis in ['bottom', 'top', 'left', 'right']:
  ax.spines[axis].set_color('white')
  ax.spines[axis].set_linewidth(3)
  ax.spines[axis].set_zorder(5)

# Líneas de puntos
ax.vlines(x=2002, ymin=0, ymax=40, color='black', linestyle='solid', linewidth=0)
ax.vlines(x=1980, ymin=14.67, ymax=26, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=1984, ymin=16, ymax=24.67, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=1988, ymin=17, ymax=27, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=1992, ymin=19, ymax=23.71, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=1996, ymin=20.17, ymax=27.67, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2000, ymin=21.67, ymax=30, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2004, ymin=21.67, ymax=30.25, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2008, ymin=20.8, ymax=30.75, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2012, ymin=19.2, ymax=33.33, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2016, ymin=22, ymax=30.75, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2020, ymin=25, ymax=32, color='white', linestyle='solid', linewidth=6)
ax.vlines(x=2024, ymin=26.2, ymax=33.38, color='white', linestyle='solid', linewidth=6)

# Texto
fig.text(0.5,0.91, "Goals per team on Men's Handball Olympic Games (1980-2024)", ha='center', va='center', fontsize=40, color='white', weight='bold')
fig.text(0.2, 0.03, 'Source: IHF (1980-2024)', ha='center', va='center', fontsize=30, color='white')
fig.text(0.513, 0.03, 'Years', ha='center', va='center', fontsize=30, color='white', weight='bold')
fig.text(0.8, 0.03, 'Graphic: Gonzalo Martínez Embil', ha='center', va='center', fontsize=30, color='white')
fig.text(0.09, 0.5, 'Goals', rotation=90, ha='center', va='center', fontsize=30, color='white', weight='bold')

# Leyenda
legend = ax.legend(['Highest scoring team per match', 'Average scoring team per match', 'Lowest scoring team per match'], markerscale=0.7, borderaxespad=1, loc='upper left', fontsize=25, facecolor='#001432', edgecolor='white', borderpad=0.75)
for text in legend.get_texts():
    text.set_fontweight("bold")
    text.set_color("white")
    text.set_fontsize(25)
    
fig.savefig('ejemplo4.png', dpi=200, bbox_inches='tight', pad_inches=1)

plt.show()
```
